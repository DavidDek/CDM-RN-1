# Red neuronal 0 - Resultados alumnos

Supón que vas a hacer un examen de lengua mañana. Además tienes un dataset con un información de cientos de alumnos que realizaron ese examen previamente.

Como ya lo has estudiado y te lo sabes a la perfección. En tu tiemplo libre has diseñado una red neuronal que predice cual es la probabilidad de que apruebes el examen.





El código está dividido en tres partes:

**1.- Dataset**: carga el dataset con toda la información de los alumnos.

**2.- Estructura de la RN**: configura la estructura de la red neuronal.

**3.- Entrenamiento y testeo**: primero, a partir de los datos, la RN se entrena para mejorar sus outputs. Despues se testea, es decir, se comprueba cómo de bien funciona la red neuronal (ya entrenada) con nuevos datos (test_data) comparando las prediciones de la RN entrenada

¡Cuando ya tienes entrenada y testeada la RN ya puedes hacer tus propias prediciones!


```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
np.random.seed(2018)
```

### 1.- Dataset

Imputs para el entrenamiento: una tabla para 1000 estudiantes con la edad, horas de estudio y nota media.

```python
x_train = np.random.random((1000,3))
```

Imputs para el testeo: una tabla para 100 estudiantes con la edad, horas de estudio y nota media.

```python
x_test = np.random.random((100, 3))
```

Outputs para las comprobaciones: una tabla con los resultados de los 1100 (1000 de entrenamiento mas 100 de testeo).

```python
y_train = np.random.randint(2, size=(1000, 1))
y_test = np.random.randint(2, size=(100, 1))
```

### 2.- Estructura de la RN.

```python
model = Sequential()
```

Toca diseñar la red neuronal:

**Imputs->** edad, horas de estudio y nota media.




**Capa 1->** 5 neuronas.


```python
model.add(Dense(5, input_dim=3, activation='relu'))
```


**Capa 2->** 4 neuronas.

```python
model.add(Dense(4, activation='relu'))
```


**Output->** probabilidad de aprobar.

```python
model.add(Dense(1, activation='sigmoid'))
```

Lo juntamos todo y definimos otros parámetros importantes.

```python
model.compile(loss='binary_crossentropy', 
              optimizer='adam',
              metrics=['accuracy'])
```

### 3.- Entrenamiento y testeo

Ahora hay que poner la RN a entrenar, es decir, a usar los datos de entrenamiento ir ajustando sus parámetros a unos mejores que le permitan que los outputs que produce se parezcan a los outputs proporcionados.

```python
model.fit(x_train, y_train, epochs=5, batch_size=32)
```

Una vez entrenada la RN, la testeamos con nuevos datos.

```python
evaluation = model.evaluate(x_test, y_test)
```

### Predicciones

Una vez esté la red entrenada y testeada correctamente ya puede hacer nuevas predicciones.

En el caso que se plantea al principio, ésta es la parte en la que le proporcionas tus datos a la RN para que te prediga las probabilidades de aprobar.

```python
new_data = np.random.random((1,3))

predictions = model.predict(new_data)

predictions
```
