# Нейронна мережа 0 - Результати учнів

Припустимо, що завтра ви збираєтеся складати мовний тест. У вас також є набір даних із інформацією про сотні студентів, які раніше складали цей іспит.

Оскільки ви його вже вивчили і знаєте досконало, у вільний час ви розробили нейронну мережу, яка прогнозує ймовірність того, що ви здасте іспит.


Код розділений на три частини:

**1.- Набір даних**: завантажте набір даних усією інформацією про студентів.

**2.- Структура RN**: налаштовує структуру нейронної мережі.

**3.- Навчання та тестування**: по-перше, на основі даних RN навчається покращувати свої результати. Потім тестується, тобто перевіряється, наскільки добре нейронна мережа (вже навчена) працює з новими даними (test_data), порівнюючи прогнози навченої RN

Коли ви вже навчилися та випробували RN, тепер можете робити власні прогнози!


```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
np.random.seed(2018)
```

### 1.- Набір даних

Вхідні дані для тренінгу: таблиця на 1000 студентів із зазначенням віку, годин навчання та середнього бала.

```python
x_train = np.random.random((1000,3))
```

Вхідні дані для тесту: таблиця на 100 учнів із зазначенням віку, годин навчання та середньої оцінки.

```python
x_test = np.random.random((100, 3))
```

Вихідні дані для тестів: таблиця з результатами 1100 (1000 навчання плюс 100 тестування).

```python
y_train = np.random.randint(2, size=(1000, 1))
y_test = np.random.randint(2, size=(100, 1))
```

### 2.- Структура НР.

```python
model = Sequential()
```

Тепер нам потрібно спроектувати нейронну мережу:

**Вхідні дані->** вік, години навчання та середній бал.




**Шар 1->** 5 нейронів.


```python
model.add(Dense(5, input_dim=3, activation='relu'))
```


**Шар 2->** 4 нейрони.

```python
model.add(Dense(4, activation='relu'))
```


**Вихід->** ймовірність проходження.

```python
model.add(Dense(1, activation='sigmoid'))
```

Ми зібрали це все разом і визначили інші важливі параметри.

```python
model.compile(loss='binary_crossentropy', 
              optimizer='adam',
              metrics=['accuracy'])
```

### 3.- Навчання та тестування

Тепер ми повинні запустити RN для навчання, тобто використовувати навчальні дані, щоб налаштувати його параметри на кращі, щоб дозволити виходам, які він виробляє, нагадувати надані вихідні дані.

```python
model.fit(x_train, y_train, epochs=5, batch_size=32)
```

Після навчання RN ми тестуємо його з новими даними.

```python
evaluation = model.evaluate(x_test, y_test)
```

### Передбачення

Після того, як мережа буде правильно навчена та перевірена, вона може робити нові прогнози.

У випадку, який виникає на початку, це частина, в якій ви надаєте свої дані RN, щоб він передбачав ймовірності проходження.

```python
new_data = np.random.random((1,3))

predictions = model.predict(new_data)

predictions
```
